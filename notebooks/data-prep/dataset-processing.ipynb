# CoLA dataset.

## Reading both .tsv files.

```python
import csv

with open('cola/in_domain_dev.tsv') as file:
    tsv_file = csv.reader(file, delimiter='\t')

    data_in_domain_dev = []

    for line in tsv_file:
        #Reading the data
        data_in_domain_dev.append(line[-1])

data_in_domain_dev[0:10]
```

```python
with open('cola/in_domain_train.tsv') as file:
    tsv_file = csv.reader(file, delimiter='\t')

    data_in_domain_train = []

    for line in tsv_file:
        #Reading the data
        data_in_domain_train.append(line[-1])

data_in_domain_train[0:10]
```

## Special character removal and tokenization.

```python
import re
import nltk

#List all word tokens removing special characters (didn't --> didnt)
word_tokens = nltk.tokenize.word_tokenize(" ".join([re.sub(r'[^a-zA-Z0-9 \n]', '', line) for line in data_in_domain_dev]))

#New token list without stopwords and punctuation symbols
filtered_tokens_dev = [word.lower() for word in word_tokens if word not in ['.', ',', '?', '!']]

filtered_tokens_dev[0:10]
```

```python
#List all word tokens removing special characters (didn't --> didnt)
word_tokens = nltk.tokenize.word_tokenize(" ".join([re.sub(r'[^a-zA-Z0-9 \n]', '', line) for line in data_in_domain_train]))

#New token list without stopwords and punctuation symbols
filtered_tokens_train = [word.lower() for word in word_tokens if word not in ['.', ',', '?', '!']]

filtered_tokens_train[0:10]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

tokenizer.fit_on_texts(filtered_tokens_dev)

tokenizer.get_config()['word_index']
```

```python
tokenizer_2 = Tokenizer()

tokenizer_2.fit_on_texts(filtered_tokens_train)

tokenizer_2.get_config()['word_index']
```

## Building the numeric representation.

```python
def text_to_token_list(word_dic: dict, text: list) -> list:
    '''
    Function that turns a text list into a token list, example:
    ["Hello",  "world", "I",  "am",  "a",  "sentence"] --> [0, 1, 3, 4, 2, 5]

    Parameters:
    word_dic: A dictionary that relates every word within the text with a numeric value, i.e. {"Hello": 0, "world": 1, "a": 2, "I": 3, "am": 4, "sentence": 5}.
    text: A list containing the text to tokenize.

    Outputs:
    token_list: A list that contains every word in the text converted into its corresponding numeric representation.
    '''
    if type(text) == list:
        return [word_dic[word] for word in text]
    else:
        return [word_dic[word] for word in text.split()]
```

```python
#Converting the word list into a token_list
token_list_dev = text_to_token_list(tokenizer.word_index, filtered_tokens_dev)

token_list_dev[0:10]
```

```python
#Converting the word list into a token_list
token_list_train = text_to_token_list(tokenizer_2.word_index, filtered_tokens_train)

token_list_train[0:10]
```

## Utility functions.

```python
def read_data(path: str) -> list:
    '''
    Function that reads a .txt file from a given path and returns a list with each sentence within the file.

    Parameters:
    path: A string containing the path of the .txt file to read.

    Outputs:
    A list containing every sentence within the .txt file.
    '''
    
    with open(path, 'r', encoding='utf-8') as file: 
        return [line.strip() for line in file]
```
